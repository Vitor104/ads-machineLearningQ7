Você trabalha para um supermercado e deseja implementar um sistema de recomendação baseado
no histórico de compras dos clientes.
Tarefas:
- Utilize um dataset de transações de supermercado contendo itens comprados em conjunto.
- Aplique o algoritmo Apriori para descobrir padrões de associação entre produtos.
- Defina regras de associação com base em suporte, confiança e lift.
- Interprete os resultados e gere recomendações de produtos.

Pergunta: Quais foram as regras de associação mais relevantes? Como elas podem ser aplicadas
para aumentar as vendas

# 1. Importar as bibliotecas necessárias

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

# 2. Carregar o dataset

```python
df = pd.read_csv('Groceries_dataset.csv')
```

```python
df.head()
```

# 3. Agrupar por Member_number e agregar os itens às listas

```python
transactions = df.groupby('Member_number')['itemDescription'].apply(list).reset_index()

# Encode as transações
te = TransactionEncoder()
te_ary = te.fit(transactions['itemDescription']).transform(transactions['itemDescription'])
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
print(df_encoded)
```

# 4. Aplicar o algoritmo Apriori

```python
frequent_itemsets = apriori(df_encoded, min_support=0.05, use_colnames=True)

print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
regras = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
print("--- Regras de Associação Encontradas ---")
display(regras[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```

# Quais foram as regras de associação mais relevantes?

R: A análise revela um padrão consistente onde a compra de diversos produtos atua como um gatilho para a compra de whole milk.

# Como elas podem ser aplicadas para aumentar as vendas?

R: As regras de associação nos permitem estar um passo à frente. Em vez de esperar que o cliente se lembre do leite, podemos usar estas informações para lembrá-lo no momento certo e da forma certa.
